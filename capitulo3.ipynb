```python
import yfinance as yf
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
import statistics as st
from scipy.stats import norm
```

```python
yf.pdr_override()
```

```python
data = pdr.get_data_yahoo('^BVSP', start='2020-01-01')
```

```python
data = data['Adj Close']
```

```python
data_vector = np.array(data)
```

```python
lin = len(data_vector)
```

```python
retorno = (data_vector[1:lin] - data_vector[0:lin - 1]) / data_vector[0:lin-1]
```

```python
grid = plt.GridSpec(2, 2, wspace=0.2, hspace=0.3)
plt.figure(figsize=(10,6))
plt.subplot(grid[0,0:])
plt.title("IBOVESPA Pontos Retorno e Distribuição")
plt.plot(data, color='gray')
plt.title('Preço')
plt.subplot(grid[1,0])
plt.plot(retorno, color='gray');
plt.title('Retorno')
plt.subplot(grid[1,1])
plt.hist(retorno, bins=40, color='gray', density=True);
xmin, xmax = plt.xlim()
media = st.mean(retorno)
desvio = st.pstdev(retorno)
eixo_x = np.linspace(xmin, xmax, 100)
eixo_y = norm.pdf(eixo_x, media, desvio)
plt.plot(eixo_x, eixo_y, color='black')
plt.title('Distribuição retorno');
```
